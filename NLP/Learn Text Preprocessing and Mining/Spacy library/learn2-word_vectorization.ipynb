Word Vectorization merupakan proses mengubah kata menjadi sebuah vektor/matriks yang berisi angka yang disebut word vector. Word vector ini merepresentasikan arti dari kata tersebut dengan angka-angka yang hanya bisa dimengerti komputer. Caranya komputer mendapatkan nilai vektor sedemikian rupa yaitu dengan cara melihat hubungan antar kata tersebut dengan kata lain dalam suatu kalimat atau seluruh dokumen.

Dalam model spacy en_core_web_md objek nlp secara otomatis melakukan word vectorization dan kita bisa mendapatkan word vector melalui atribut .vector

```python
import spacy

nlp= spacy.load('en_core_web_md')
```

```python
with open ('cinderella.txt', 'r') as f:
    text= f.read()
```

```python
doc= nlp(text)

list(doc.sents)[0]
```

Mendapatkan vektor

```python
sent1= list(doc.sents)[0]

sent1.vector
```
